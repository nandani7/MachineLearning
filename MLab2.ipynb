```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('share.csv')
```

```python
data.head()
```

```python
data.replace(['*','**'],np.nan,inplace=True)
```

```python
data
```

```python
data['Upper 95% Confidence Interval for Trend']=data['Upper 95% Confidence Interval for Trend'].apply(lambda x:float(x))
data['Upper 95% Confidence Interval for Trend'].replace(np.nan,data['Upper 95% Confidence Interval for Trend'].mean(),inplace=True)
data['Upper 95% Confidence Interval for Trend']
```

```python
data['Lower 95% Confidence Interval for Death Rate']=data['Lower 95% Confidence Interval for Death Rate'].apply(lambda x:float(x))
data['Lower 95% Confidence Interval for Death Rate'].replace(np.nan,data['Lower 95% Confidence Interval for Death Rate'].mean(),inplace=True)
data['Lower 95% Confidence Interval for Death Rate']
```

```python
data['Recent 5-Year Trend (2) in Death Rates']=data['Recent 5-Year Trend (2) in Death Rates'].apply(lambda x:float(x))
data['Recent 5-Year Trend (2) in Death Rates'].replace(np.nan,data['Recent 5-Year Trend (2) in Death Rates'].mean(),inplace=True)
data['Recent 5-Year Trend (2) in Death Rates']
```

```python
data
```

```python
data['Lower 95% Confidence Interval for Trend']=data['Lower 95% Confidence Interval for Trend'].apply(lambda x:float(x))
data['Lower 95% Confidence Interval for Trend'].replace(np.nan,data['Lower 95% Confidence Interval for Trend'].mean(),inplace=True)
data['Lower 95% Confidence Interval for Trend']
```

```python
data
```

```python
x=data['Recent 5-Year Trend (2) in Death Rates']
y=data['Lower 95% Confidence Interval for Trend']
plt.scatter(x,y,marker='.',label='Scatter plot')
min_x=min(x)-100
max_x=max(x)+100
X=np.linspace(max_x,min_x,1000)
Y=X
plt.plot(X,Y,color='red',label='Regression Line')
plt.xlim([-60,60])
plt.ylim([-60,60])
plt.legend()
plt.xlabel("Recent 5-Year Trend (2) in Death Rates")
plt.ylabel("Lower 95% Confidence Interval for Trend")
```

```python
x=data['Lower 95% Confidence Interval for Trend']
y=data['Upper 95% Confidence Interval for Trend']
plt.scatter(x,y,marker='.',label='Scatter plot')
min_x=min(x)-100
max_x=max(x)+100
X=np.linspace(max_x,min_x,1000)
Y=X
plt.plot(X,Y,color='red',label='Regression Line')
plt.xlim([-60,60])
plt.ylim([-60,60])
plt.legend()
plt.xlabel("Recent 5-Year Trend (2) in Death Rates")
plt.ylabel("Lower 95% Confidence Interval for Trend")
```
